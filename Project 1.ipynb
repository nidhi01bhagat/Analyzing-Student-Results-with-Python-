# gender distribution 

```python
import pandas as pd
import seaborn as sns  # Import the seaborn library
import matplotlib.pyplot as plt  # Import matplotlib for plotting
df = pd.read_csv('Expanded_data_with_more_features.csv')
plt.figure(figsize= (6,6))#reduce and increase the height of the chart
# Assuming df is already defined and loaded with data
ax = sns.countplot(data=df, x="Gender")
ax.bar_label(ax.containers[0])   

plt.show()  # Display the plot
```

# from the above chart we have analysed that:
#the number of females in the data is more than the number of males 

```python
import pandas as pd
df = pd.read_csv('Expanded_data_with_more_features.csv')
gb = df.groupby("ParentEduc").agg({"MathScore": 'mean',"ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 
plt.figure(figsize= (4,4))
sns.heatmap(gb,annot = True)#annot is used for showing the value in the heat map
plt.title("Gender Distribution")
plt.show()
```

```python

#from the above chart we have concluded that the education of the parents have a good impact on the scores. 
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore": 'mean',"ReadingScore":"mean","WritingScore":"mean"})
print(gb1)
```

```python
plt.figure(figsize= (4,4))
sns.heatmap(gb1,annot = True)#annot is used for showing the value in the heat map
plt.title("Relationaship Between Parent's Education and Student Score")
plt.show()
```

```python
#from the above chart we have concluded that there is no/negligible impact on the student's score due to their parent's marital status 
```

# Distribution of Ethnic Groups

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('Expanded_data_with_more_features.csv')
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()

l = ["group A", "group B", "group C", "group D", "group E",]
mlist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
print(mlist)
plt.pie(mlist, labels = l, autopct = "%1.2f%%") 
plt.title("Distribution of Ethnic Groups")
plt.show()
```
